```python
# Задание № 1
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')

join = ratings.merge(movies, on = 'movieId', how = 'left')
rating_5_films = join[join['rating'] >= 5]
rating_5_films['title'].value_counts().head(10)
```

```python
# Задание № 2

power = pd.read_csv('power.csv')
# Фильтр искомых стран
baltic_countries = power[(power['country'] == 'Estonia') | (power['country'] == 'Lithuania') | (power['country'] == 'Latvia')]
# Фильтр искомых категорий
baltic_categories = baltic_countries[(baltic_countries['category'] == 4) | (baltic_countries['category'] == 12) | (baltic_countries['category'] == 21) ]
# Фильтр по годам
year_filter = baltic_categories[(baltic_categories['year'] <= 2010) & (baltic_categories['year'] >= 2005) ]
# Исключение отрицательного потребеления + подсчет
final_result = year_filter[year_filter.quantity >0].quantity.sum()

final_result
```
