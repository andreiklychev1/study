```python
import requests

class Rate:
    
    def __init__(self, format_='value'):
        self.format = format_
        
    
    def exchange_rates(self):

        self.r = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')
        return self.r.json()['Valute']
    
    def make_format(self, currency):
  
        
        response = self.exchange_rates()
        
        
        if currency in response:
            if self.format == 'full':
                return response[currency]
            
            if self.format == 'value':
                
                return response[currency]['Value']
        
            
        return 'Error'
# Задание № 2   
    def difference(self, currency):
        
        response = self.exchange_rates()
        
        if currency in response:
            
            if self.format == 'value':
                return round(response[currency]['Value'] - response[currency]['Previous'], 2)
        
        return 'Error'
        
        
    def eur(self, diff = False):
        
        self.diff = diff
    
        if diff == False:
            return self.make_format('EUR')
        elif diff:
            return self.difference('EUR')
                
    def usd(self, diff = False):
        self.diff = diff
      
        if self.diff == False:
            return self.make_format('USD')
        elif self.diff:
            return self.difference('USD')
# Задание № 1        
    
    def max_(self):
        dictionaries = self.exchange_rates().values()
        
        val = []
        for element in dictionaries:
            val.append(element["Value"])
        
        for element in dictionaries:
            if max(val) in element.values():
                return element['Name'], element["Value"] 
                break
        
        
        
```
