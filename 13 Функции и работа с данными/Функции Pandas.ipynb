```python
# Задание №1
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
join = ratings.merge(movies, on = 'movieId', how = 'left')

def classificator(table):
    
    if table['rating'] <= 2:
        return 'низкий рейтинг'
    elif 2 < table['rating'] < 4.5:
        return 'средний рейтинг'
    elif 4.5 <= table['rating'] <= 5:
        return 'высокий рейтинг' 

join['class'] = join.apply(classificator, axis = 1)
join
```

```python
# Задание №2
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
queries = pd.read_csv('keywords.csv')
def counter(table):
    cities = []
    for element in geo_data:
        cities.extend(geo_data[element])
    for el in enumerate(cities):
        if el[1] in table['keyword']:
            if 0 <= el[0] <= 2:
                return 'Центр'
            elif 3 <= el[0] <= 5:
                return 'Северо-Запад'
            elif 6 <= el[0] <= 8:
                return 'Дальний Восток'
            
    else:
        return 'undefined'

```

```python
queries['region'] = queries.apply(counter, axis=1)
```

```python
# Задание №3
years = []
year = 1950
while year != 2011:
    years.append(year)
    year += 1
def production_year(table):
    for element in years:
        if str(element) in table.title:
            return element
    else:
        return '1900'   
    
```

```python
join['year'] = join.apply(production_year, axis=1) 
```
