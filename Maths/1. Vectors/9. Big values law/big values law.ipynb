Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.

```python
from scipy import special
```

```python
def rand_gamma(alpha = 2, lambda_ = 2):
    x = np.random.random(1)
    rv = alpha**(lambda_)/special.gamma(lambda_) * x**(lambda_ - 1) * np.exp(-alpha*x)
    return rv

# создадим выборку из 1000 значений
sample = [rand_gamma()[0] for _ in range(1000)]
```

```python

sns.distplot(sample)

plt.grid()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
# Параметры запуска: размеры выборок и количество повторов генерации

# YOUR CODE HERE
def CPT(n1):
    samples_count = 1000
    # Массивы для сохранения средних по выборкам
    a1 = np.array([])

    for number in range(samples_count):
        # На каждом цикле генерируем выборки разных размеров
        sample1 = [rand_gamma()[0] for _ in range(n1)]
        # YOUR CODE HERE
        # Считаем среднее по каждой выборке и добавляем в массив средних
        a1 = np.append(a1, sum(sample1)/len(sample1))
        # YOUR CODE HERE
    return a1
```

```python
# Строим графики
# Подписи для графиков
label_1 = 'sample of ' + str(n1) + ' histogram'
sns.distplot(CPT(2))
sns.distplot(CPT(5))
sns.distplot(CPT(10))
sns.distplot(CPT(50))
sns.distplot(CPT(100))

plt.legend(title='Кол-во знач-ий в выборке', labels=['n=2', 'n=5', 'n=10', 'n=50', 'n=100'])
plt.grid()
plt.ylabel('samples means')
plt.xlabel('$x$')
plt.show()
```
