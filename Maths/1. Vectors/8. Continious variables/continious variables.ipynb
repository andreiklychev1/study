### Задание 1

Нормальное распределение

Функция плотности распределния зависит от 2-ух параметров - мат. ожидание и стандартное отклонение  
Для наглядности изменений будем менять парметр стандартное отклонение

```python
def values(x=np.random.randint(1), mu=np.random.randint(1)):
    return np.random.normal(x, mu, 1000)

def m_expect(X=values()):
    return X.mean()



def disp(X=values(), E=m_expect()):
    d = 0
    for _ in X:
        d += (E - _)**2
    return d/len(X)

def f_density(X=values(), E=m_expect()):
    array = []
    for i in X:
        
        array.append(1/(2*np.pi*disp(X, E))**(1/2) * np.exp(-(E-i)**2 / disp(X, E)))
    return array


X1 = values(2, 2)
y1 = f_density(X=X1)

X2 = values(2, 3)
y2 = f_density(X=X2)

X3 = values(2, 5)
y3 = f_density(X=X3)

X4 = values(2, 10)
y4 = f_density(X=X4)

X5 = values(2, 15)
y5 = f_density(X=X5)
```

```python
sns.lineplot(X1, y1)
sns.lineplot(X2, y2)
sns.lineplot(X3, y3)
sns.lineplot(X4, y4)
sns.lineplot(X5, y5)
plt.legend(title='STDEV', labels=['mu=2', 'mu=3', 'mu=5', 'mu=10', 'mu=15'], fontsize=12)

plt.grid()
plt.show()
```

Экспоненциальное распределение

Будем изменять в данном опыте параметр лямбда

```python
x = [_ for _ in range(100)]

def exp(lambda_):
    array = []
    for k in x:
        y = lambda_*np.exp(-0.25*k)
        array.append(y)
    return array

y1 = exp(0.25)
y2 = exp(0.5)
y3 = exp(1)
y4 = exp(1.5)
y5 = exp(2)
```

```python
sns.lineplot(x, y1)
sns.lineplot(x, y2)
sns.lineplot(x, y3)
sns.lineplot(x, y4)
sns.lineplot(x, y5)
plt.legend(title='lambda', labels=['l=0.25', 'l=0.5', 'l=1', 'l=1.5', 'l=2'], fontsize=15)
plt.grid()
plt.show()
```

Распределение Стьюдента

Здесь основной параметр - кол-во значений в выборке  
Значения должны быть из выборки распредленной по нормальному закону
Чем больше значений мы имеем, тем больше наше распределение будет походить на нормальное

```python
from scipy import special
from scipy import stats as sts
```

```python

def student(m):
    x = np.linspace(-10, 10, m)
    return sts.t(m).pdf(x), x

x1 = student(5)[1]
y1 = student(5)[0]

x2 = student(10)[1]
y2 = student(10)[0]

x3 = student(20)[1]
y3 = student(20)[0]

x4 = student(50)[1]
y4 = student(50)[0]

x5 = student(100)[1]
y5 = student(100)[0]
```

```python
sns.lineplot(x1, y1)
sns.lineplot(x2, y2)
sns.lineplot(x3, y3)
sns.lineplot(x4, y4)
sns.lineplot(x5, y5)
plt.legend(title='n', labels=['n=5', 'n=10', 'n=20', 'n=50', 'n=100'], fontsize=15)
plt.grid()
plt.show()
```
