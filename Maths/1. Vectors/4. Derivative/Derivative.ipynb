```python
import math
import numpy as np
from scipy import misc
```

### Уровень 0

```python
def func(x):
    return x**5 + 4*sin(2*x) + cos(3*x + 3)
def first_dev(x):
    return 5*x**4 + 8*cos(2*x) - 3*sin(3*x + 3)
```

Первая производная: 5*x**4 + 8*cos(2*x) - 3*sin(3*x + 3)  
Вторая производная: 20*x**3 - 16*sin(2*x) - 9*cos(3*x + 3)

```python
# Значение первой производной в точке 1
misc.derivative(func, 1, dx=1e-6)
```

```python
# Значение второй производной в точке 1
misc.derivative(first_dev, 1, dx=1e-6)
```

Посчитаем обе производные программным способом:

```python
from sympy import *
x = Symbol('x')
y = func(x)
first_derivative = y.diff(x)
first_derivative
```

```python
y = first_derivative
second_derivative = y.diff(x)
second_derivative
```

```python
der_1 = lambdify(x, first_derivative)
der_2 = lambdify(x, second_derivative)
```

```python
print('Значение первой производной в точке 1 = ', der_1(1))
print('Значение второй производной в точке 1 = ', der_2(1))
```

### Уровень 1

```python
import matplotlib
from matplotlib import pyplot as plt
matplotlib.rcParams['figure.figsize'] = [10, 7]
```

```python
def f(x):
    return (sin(2*x + 1))**5
```

```python
first_der = lambdify(x, f(x).diff(x))
y = f(x).diff(x)
second_der = lambdify(x, y.diff(x))
```

```python
x = np.linspace(-5, 5, 50)
y1 = []
y2 = []
for _ in range(len(x)):
    y1.append(first_der(x[_]))
    y2.append(second_der(x[_]))
```

```python
plt.plot(x, y1)
plt.ylabel('Значения производной')
plt.xlabel('Пределы исследования')
plt.grid()
plt.show()
```

```python
import seaborn as sns
```

```python
sns.lineplot(x, y2)
plt.ylabel('Значения второй производной')
plt.xlabel('Пределы исследования')
plt.grid()
plt.show()
```
