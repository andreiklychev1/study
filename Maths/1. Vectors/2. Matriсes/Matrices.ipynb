Уровень 1

```python
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]
A = np.array(A)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.grid()
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

### Масштабирование объекта в 0.5 раз

```python
scale_matrix = (1/2) * np.eye(3)
scale_matrix
```

```python
B = [A[i] @ scale_matrix for i in range(0, len(A))]
B = np.concatenate(B, axis=0).reshape(6, 3)
```

```python

x = B[:,0]
y = B[:,1]
plt.plot(x, y)
plt.grid()
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

### Поворот объекта на 130 градусов

```python
theta = np.pi * 130 /360
turning_matrix = np.array([[np.cos(theta), -np.sin(theta), 0],
                         [np.sin(theta), np.cos(theta), 0],
                         [0, 0, 1]])
```

```python
C = [A[i] @ turning_matrix for i in range(0, len(A))]
C = np.concatenate(C, axis=0).reshape(6, 3)
```

```python
x = C[:,0]
y = C[:,1]
plt.plot(x, y)
plt.grid()
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

### Отображение относительно прямой y=x

```python
reflection_matrix = np.array([[0, 1, 0],
                             [1, 0, 0],
                             [0, 0, 1]])
```

```python
D = [A[i] @ reflection_matrix for i in range(0, len(A))]
D = np.concatenate(D, axis=0).reshape(6, 3)
```

```python
x = D[:,0]
y = D[:,1]
plt.plot(x, y)
plt.plot((-150, 150), (-150, 150), 'r')
plt.grid()
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

### Смещение и масштабирование объекта

```python
movement_matrix = np.array([[1, 0, 200],
                           [0, 1, 300],
                           [0, 0, 1]])

scale_matrix_2 = np.array([[0.5, 0, 0],
                          [0, 1.2, 0],
                           [0, 0, 1]])

transformation_matrix = movement_matrix @ scale_matrix_2
transformation_matrix
```

```python
E = [A[i] @ transformation_matrix for i in range(0, len(A))]
E = np.concatenate(E, axis=0).reshape(6, 3)
```

```python
x = E[:,0]
y = E[:,1]
plt.plot(x, y)
plt.grid()
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```
